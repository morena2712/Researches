### Analisi dell'attendibilità

Il passo successivo è stato calcolare le α di Cronbach per ogni test e confrontarle. Per farlo, c’è stato bisogno di scaricare il pacchetto `cocron` per R dal sito cocron - comparing Cronbach alphas. http://comparingcronbachalphas.org/. Questo
pacchetto permette di comparare n coefficienti α dipendenti o indipendenti
(Cronbach, 1951), con l’implementazione dei metodi di Feldt, Woodruff e Salih
(1987). Di seguito i comandi per calcolare le alpha:

```r
print("L'α di Cronbach del BAQ è:")
cocron::cronbach.alpha(CO)
```

```r
print("L'α di Cronbach della Versione 1 è:")
cocron::cronbach.alpha(C1)
```

```r
print("L'α di Cronbach della Verisone 2 è:")
cocron::cronbach.alpha(C2)
```

In seguito, si sono confrontati i punteggi di tutti e tre i test. Si è prima creato un
vettore che contenesse le α, poi un vettore che contenesse le numerosità dei
sottogruppi, e infine si è eseguito il confronto. Di seguito i comandi per confrontare
tutti i test tra loro:

```r
alpha<-c(0.775, 0.817, 0.840)
n<-c(57, 65, 56)
cocron::cocron.n.coefficients(alpha, n, items = 18,
dep = FALSE, los = 0.10, conf.level = 0.90)
```

Questo confronto ha riportato un valore del p-value uguale a 0.47. Nonostante
non si sia rilevata una differenza significativa tra i test, si è proceduto a effettuare
l’analisi tra coppie di test, in modo da avere un’informazione più dettagliata.\
Anche in questo caso, si sono prima creati dei vettori contenenti le alpha dei test
confrontati e dei vettori con la numerosità dei sottogruppi. Di seguito i comandi:

```r
alphaC1C2<-c(0.775, 0.817)
alphaC2CO<-c(0.817,0.840)
alphaC1CO<-c(0.775,0.840)
n12<-c(57,65)
n2O<-c(65, 56)
n1O<-c(57,56)

print("Confronto tra Versione 1 e Versione 2:")
cocron::cocron.two.coefficients(alphaC1C2, n12, dep
= FALSE, los = 0.05, alternative = "two.sided")
```

```r
print("Confronto tra Versione 2 e BAQ:")
cocron::cocron.two.coefficients(alphaC2CO, n2O, dep
= FALSE, los = 0.05, alternative = "two.sided")
```

```r
print("Confronto tra Versione 1 e BAQ:")
cocron::cocron.two.coefficients(alphaC1CO, n1O, dep
= FALSE, los = 0.05, alternative = "two.sided")
```

I risultati ottenuti confermano che non ci sono differenze significative tra i test.

### Analisi della validità
Successivamente si è svolta l’Analisi in Componenti Principali per condurre una
valutazione esplorativa della validità. È stato necessario installare il pacchetto
`FactoMiner` su RStudio. Con il seguente script, è stato possibile avere un’analisi
completa con pochi comandi:

```r
library(FactoMineR)
res.pca1=PCA(C1)
summary(res.pca1)
lapply(dimdesc(res.pca1),lapply,round,6)
head(res.pca1$eig)
round(res.pca1$var$contrib,3)
round(res.pca1$var$cos2,3)
```

```r
plot(res.pca1,choix="var")
```

```r
barplot(res.pca1$eig[,1], main="Scree-plot degli autovalori (Versione 1)", col = "springgreen3")
```

```r
res.pca2=PCA(C2)
summary(res.pca2)
lapply(dimdesc(res.pca2),lapply,round,6)
head(res.pca2$eig)
round(res.pca2$var$contrib,3)
round(res.pca2$var$cos2,3)
```

```r
plot(res.pca2,choix="var")
```

```r
barplot(res.pca2$eig[,1],main="Scree-plot degli autovalori (Versione 2)", col = "purple")
```

```r
res.pcaO=PCA(CO)
summary(res.pcaO)
lapply(dimdesc(res.pcaO),lapply,round,6)
head(res.pcaO$eig)
round(res.pcaO$var$contrib,3)
round(res.pcaO$var$cos2,3)
```

```r
plot(res.pcaO,choix="var")
```

```r
barplot(res.pcaO$eig[,1], main="Scree-plot degli autovalori (BAQ)", col = "pink")
```

Il comando `library` richiama l’utilizzo del pacchetto `FactoMiner`. Il
comando `PCA` svolge l’Analisi in Componenti Principali riducendo la
dimensionalità dei dati con una minima perdita di informazione, permettendo
anche di visualizzarli graficamente. Con `summary` si produce un report
sull’importanza delle componenti in termini di deviazione standard (sd) e
proporzione della varianza spiegata e della varianza spiegata cumulata sia per gli
individui che per le variabili. `lapply` restituisce la misura delle correlazioni e del
p-value, mentre `dimdesc` fornisce una descrizione che precisa quali variabili siano
maggiormente caratteristiche secondo ciascuna dimensione. Si specifica `lapply`
in modo da applicare tali funzioni a un elenco (lista) e non un vettore, e `round`
specifica a quanti numeri dopo la virgola arrotondare. Con il comando `plot` si
disegna il cerchio delle correlazioni; con `choix` si scelgono i dati da rappresentare
(in questo caso le variabili). Il comando `round` formatta i valori, in questo caso i
contributi delle singole modalità alla varianza e i 𝑐𝑜𝑠2, arrotondati alla terza cifra
dopo la virgola. Con il comando `barplot` si disegna il diagramma a barre degli
autovalori (scree-plot), dandoci informazioni della varianza spiegata da ogni
componente.\
I risultati restituiti suggeriscono che: nel BAQ originale nove dimensioni
spiegano l’81,71% della varianza totale; nella Versione 1 nove dimensioni spiegano
l’80,07% della varianza totale; nella Versione 2 dieci dimensioni spiegano
l’83,27% della varianza totale.

### Discussione
Dai risultati sembrerebbe non esserci una differenza statisticamente significativa
tra i tre questionari.\
Nella selezione dei test creati con l’IA, il riscontro di un progressivo aumento
delle domande legate a stati emotivi o mentali (esplicitamente non richiesti nella
formulazione del prompt) potrebbe far pensare ad un’influenza delle passate
interazioni con il modello Artificiale non solo nello stesso spazio di comunicazione,
ma anche in chat diverse.\
Per quanto riguarda l’attendibilità, misurata con le α di Cronbach, tutti e tre i test
risultano attendibili; in particolare, l’alpha del BAQ originale (α = 0.84) non è
diversa da quella riscontrata in letteratura (α = 0.83). È interessante notare come
l’alpha della Versione 1 (α = 0.77) sia molto differente rispetto a quella del BAQ
originale, nonostante fosse la riformulazione che presentava il prompt con maggiori
specifiche e con domande più simili. Ciò potrebbe far pensare ad un effetto di
confusione generato dal grado di dettaglio del prompt; formulazioni più aperte,
meno dettagliate, potrebbero permettere al modello di cercare tra più dati per
giungere ad una risposta ottimale. Questa ipotesi potrebbe essere avvalorata sia dal
valore dell’alpha della Versione 2 (α = 0.82) che dal p-value ottenuto dal confronto tra le α del BAQ originale e della Versione 1, il quale, seppur non statisticamente
significativo, risulta essere il più basso tra i vari confronti (p-value = 0.21),
permettendo di ipotizzare che queste due versioni siano quelle più differenti tra loro.\
Per quanto riguarda la validità, con la PCA si ottengono per i tre questionari
risultati concordi con quella presenti in letteratura. In particolare, ci si aspetta dai
quattro ai sei fattori che spieghino la varianza dei test. Dai grafici degli scree-plot,
che ci danno informazioni sulla varianza spiegata, notiamo che il numero di
componenti con un autovalore maggiore di 1 rientra in questo range per ogni
formulazione del questionario. È importante sottolineare il fatto che si tratta di due
analisi diverse: la PCA è un’analisi esplorativa che non si fonda su specifiche
assunzioni legate a un modello strutturale né a distribuzioni statistiche soggiacenti,
mentre quella effettuata in letteratura è un’analisi fattoriale. Anche per quanto
riguarda il cerchio delle correlazioni, si nota una distribuzione simile delle variabili,
con leggere differenze nella qualità della rappresentazione (lunghezza delle frecce).
Confrontando i 𝑐𝑜𝑠2, notiamo che: gli item che meglio rappresentano la prima
dimensione nel BAQ originale sono M3 e M4, permettendoci di ipotizzare che
questa dimensione faccia riferimento all’esordio delle malattie; la seconda
dimensione è meglio rappresentata dagli item P12 e P13, facendoci pensare che
faccia riferimento al predire le reazioni corporee; la terza è meglio rappresentata
dagli item N15 e N16, riferendosi probabilmente alle reazioni corporee al cibo; la
quarta dimensione è meglio rappresentata dagli item N17 e C14, facendo forse
riferimento alla consapevolezza dei cicli di funzionamento; e la quinta dimensione
è meglio rappresentata dall’item P16, riferendosi ad aspetti legati al sonno. Nella
Versione 1, la prima dimensione è meglio rappresentata dagli item C2 e C3, facendo
probabilmente riferimento ad aspetti legati alla stanchezza; la seconda dimensione
è meglio rappresentata dagli item P1 e P2, riferendosi forse al predire le reazioni
corporee; la terza dimensione è meglio rappresentata dagli item P4 e C4, facendo
probabilmente riferimento ad aspetti legati al sonno; la quarta dimensione è meglio
rappresentata dagli item N5 e C5, riferendosi forse ai livelli di energia; la quinta
dimensione è meglio rappresentata dall’item N3, facendo riferimento ai
cambiamenti stagionali. Nella Versione 2, la prima dimensione è meglio
rappresentata dagli item P6 e C7, facendo probabilmente riferimento alla consapevolezza dei cicli di funzionamento; la seconda dimensione è meglio
rappresentata dagli item N7 e N8, riferendosi forse al rilevamento delle funzioni
corporee; la terza dimensione è meglio rappresentata dagli item P11 e N9,
riferendosi probabilmente alla percezione di malessere; la quarta dimensione è
meglio rappresentata dall’item C6, facendo riferimento ai livelli di energia; infine,
la quinta dimensione è meglio rappresentata dagli item N11 e P7, riferendosi
probabilmente al riconoscimento delle sensazioni.\
Le dimensioni trovate e la loro distribuzione sono più o meno simili tra i diversi
questionari. In conclusione, questo studio sembra non aver riscontrato differenze
significative né per l’attendibilità dei test, né per la loro validità.
### Limiti e prospettive future
Questa ricerca presenta diversi limiti. La traduzione del BAQ originale
dall’inglese all’italiano è stata effettuata dalla sottoscritta. Oltre a questo, lo studio
è stato condotto utilizzando un campione non probabilistico, costituito da una forte
presenza di soggetti femminili. Un miglior campionamento, una maggiore presenza
maschile per valutare possibili influenze del sesso biologico nella selezione delle
risposte, e una traduzione più accurata potrebbero essere impiegati in studi futuri.
Nelle prossime ricerche si potrebbe anche approfondire l’algoritmo, il modello di
interpretazione dei prompt dell’IA e il modo in cui vengono selezionati i dati
d’addestramento per spiegare il motivo per cui sono state generate domande che
non corrispondevano in maniera efficace a ciò che è stato richiesto. Si potrebbero
formulare prompt diversi, utilizzare altri modelli IA e/o altre tecniche di analisi
statistica e psicometrica per validare i risultati, come ad esempio effettuare
un’analisi fattoriale. Ancora, si potrebbe utilizzare uno strumento Intelligente per
creare un test ex novo per poi valutarne l’attendibilità e la validità convergente e
discriminante, così da misurare l’effettiva capacità generativa dell’IA piuttosto che
la sua capacità di emulazione di fonti già esistenti.

##Conclusioni
L'obiettivo della ricerca era quello di verificare l’attendibilità e la validità di due strumenti psicologici creati da Chat
GPT, mettendoli a confronto con il Body Awereness Questionnaire (BAQ), un test
già validato e presente in letteratura. La ricerca presenta diversi limiti, tra cui il fatto
di lavorare con un campione non probabilistico, la scarsa presenza di soggetti
maschili all’interno del campione, l’incapacità di sapere come il modello Artificiale
abbia selezionato le domande da presentare. Utilizzando le misure dell’α di
Cronbach, e confrontando fra loro i test, i risultati suggeriscono che non è presente
una differenza statisticamente significativa tra le tre versioni per quanto riguarda
l’attendibilità. Con l’Analisi in Componenti Principali non si sono riscontrate
differenze significative neanche per quanto riguarda la validità. Ricerche future
potrebbero concentrarsi maggiormente sulla formulazione dei prompt, selezionare
in maniera più accurata i campioni, utilizzare un altro modello Artificiale o altri test
statistici, o creare un test originale e validarlo.
