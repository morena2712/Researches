## **Analisi del Mental Health Score e del Social Media Addiction Score**

**Abstract.** Il presente studio indaga l’impatto dell’utilizzo dei social media sul benessere psicofisico
della popolazione studentesca di età compresa tra i 16 e i 25 anni, analizzando le determinanti del
punteggio di salute mentale (Mental Health Score) e del punteggio di dipendenza (Addicted Score). La
ricerca si basa sul dataset cross-sectional ”Student Social Media & Relationships” (2025), comprendente
705 osservazioni globali.
Dal punto di vista metodologico, dopo aver effettuato una selezione delle variabili tramite l'algoritmo Sequential Replacement, l’analisi ha evidenziato i limiti della regressione lineare multipla
(violazione dell’omoschedasticità e natura discreta dei dati). Di conseguenza, sono stati adottati Modelli Lineari Generalizzati (GLM) con distribuzione Quasi-Poisson e funzione di legame logaritmica, che
hanno garantito stime robuste e una corretta gestione della varianza.
I risultati indicano che il modello per la salute mentale spiega oltre il 70% della variabilità. Si
osserva una netta gerarchia del rischio legata all’architettura delle piattaforme: i social basati sul
confronto visivo e sullo scorrimento infinito (TikTok, Instagram, Snapchat) incidono negativamente sul
benessere, mentre le piattaforme di utilità e messaggistica (WhatsApp, X/Twitter) mostrano effetti
neutri o positivi. Il modello relativo alla dipendenza digitale spiega circa il 91% della variabilità. In
questo contesto, il predittore più critico risulta essere l’impatto percepito sulla performance accademica,
mentre YouTube emerge come la piattaforma a più alto rischio di utilizzo compulsivo, in contrasto con
l’effetto protettivo delle app di messaggistica diretta.
In entrambi i modelli, la qualità del sonno si conferma il principale fattore protettivo biologico. Le
conclusioni suggeriscono che la salute digitale non dipenda esclusivamente dal tempo di utilizzo, ma sia
il risultato dell’interazione tra i ritmi biologici individuali e le specifiche meccaniche di ingaggio delle
diverse piattaforme.

**1.1 *Bergen Social Media Addiction Scale (BSMAS)***

La Bergen Social Media Addiction Scale è utilizzata nella misurazione della dipendenza da social media.
Questo strumento si basa sul modello biopsicosociale delle dipendenze teorizzato da Griffiths.
Questo modello, originariamente sviluppato per le dipendenze comportamentali, sostiene che tutte
le forme di dipendenza condividono un nucleo comune di sei componenti: Salienza, Modificazioni
dell’umore, Tolleranza, Astinenza, Conflitto, Ricaduta. Queste componenti sono dei veri e propri
criteri strutturali: se un comportamento non mostra questo pattern, allora è più corretto parlare di
”uso problematico” piuttosto che di ”dipendenza”.
Il test è composto da sei affermazioni riferite alle esperienze avute nell’ultimo anno, ciascuna valutata su scala Likert da 1 a 5. Il risultato si calcola sommando il punteggio degli item; valori più alti
indicano maggiore rischio di dipendenza dai social media.
Il test è stato ampiamente validato e accettato da meta-analisi e studi psicometrici. Tuttavia,
alcuni autori sostengono che applicare il modello di addiction ai social media rischi di patologizzare
i comportamenti attuati in contesti in cui l’uso intenso dei social è socialmente richiesto, e di
confondere l’alta frequenza d’uso con la dipendenza senza distinguere chiaramente tra un uso intenso
ma funzionale e un uso disfunzionale. Esiste una controversia riguardo al punteggio di cut-off
ottimale per definire la dipendenza; alcuni studi propongono una soglia di 19 su 30, mentre ricerche
basate su interviste cliniche suggeriscono un cut-off di 24 o 26.
Infine, il modello di Griffiths non è stato sviluppato tenendo in considerazione le specificità dei
social, di conseguenza potrebbe non cogliere alcune dimensioni specifiche come ad esempio le dinamiche
di notifica, il confronto sociale, la ricerca di approvazione e la FOMO.
Si è discusso della struttura di questo test poiché i dati che verranno analizzati sono stati raccolti
attraverso un sondaggio online i cui item sono stati adattati a partire da quelli di scale validate, in
particolare la BSMAS.

L’analisi dei punteggi BSMAS rivela come l’interazione con le piattaforme digitali stia evolvendo da
abitudine a potenziale patologia. Questi punteggi non seguono una distribuzione normale perfetta,
ma tendono a raggrupparsi verso la fascia bassa e media, con una coda lunga verso i punteggi alti che
indicano la dipendenza. In un campione di studenti universitari malesi, il punteggio medio totale alla
BSMAS è risultato essere 16.80 (SD = 5.40) su un massimo di 30. Questo valore medio colloca
la popolazione studentesca in una zona di ”rischio moderato” o uso intensivo, ma non necessariamente patologico. Tuttavia, la deviazione standard indica che una quota rilevante del campione ottiene
punteggi superiori a 22-24, entrando nella fascia di rilevanza clinica. La prevalenza stimata della
dipendenza varia enormemente in base al cut-off adottato.
I punteggi di dipendenza non sono uniformi a livello globale. Studi comparativi mostrano differenze
significative tra culture individualiste e collettiviste. In Italia, gli studenti mostrano livelli particolarmente elevati di ”Fear of Missing Out” (FOMO) e di dipendenza da internet in generale, ma punteggi
di dipendenza dai social media (SMA) che possono essere inferiori rispetto ai coetanei di altri paesi
europei o sudamericani, pur mantenendo alti livelli di phubbing (ignorare gli altri per guardare il

un imperativo culturale, i punteggi possono essere influenzati dalla pressione normativa a essere sempre
reperibili. Tuttavia, la forte coesione familiare in paesi come l’Indonesia può fungere da fattore
protettivo, moderando l’impatto negativo dello stress pandemico sull’uso dei social.

**1.4. Variabili intervenienti nel fenomeno e influenza dei contesti sociodemografici**

La correlazione tra punteggi di salute mentale scarsi e alti punteggi di dipendenza dai social media è
robusta e ben documentata; tuttavia, la ricerca psicologica ha indagato le variabili intervenienti
(mediatori e moderatori) che potrebbero spiegare il fenomeno.
La FOMO emerge come il predittore più potente nella relazione tra disagio psicologico e SMA.
Questa è una forma d’ansia sociale caratterizzata dal timore pervasivo che gli altri stiano vivendo esperienze gratificanti in propria assenza. Analisi di Structural Equation Modeling (SEM) confermano
che la FOMO media la relazione tra psicopatologia e uso problematico dei social media. Individui
con bisogni psicologici insoddisfatti (es. bisogno di appartenenza) sviluppano la FOMO, che a sua volta
guida comportamenti di controllo compulsivo delle notifiche, risultando in punteggi SMA più elevati.
Un altro mediatore della relazione tra dipendenza dai social e salute mentale è il confronto sociale.
L’esposizione costante a rappresentazioni idealizzate e curate della vita altrui innesca confronti verso
l’alto, provocando sentimenti di invidia, inadeguatezza e, conseguentemente, portando a punteggi più
alti di depressione. A questa evidenza si associano quelle riferite alle modalità di utilizzo dei
social. Un uso passivo (scrolling), senza interazione diretta, è costantemente associato a un aumento
del malessere, probabilmente per via della mancanza di feedback sociali positivi e all’aumento delle
opportunità di confronto sociale. Un uso attivo, come messaggiare o commentare, mostra invece
risultati ambivalenti; da un lato sembra favorire la connessione sociale, che è un fattore protettivo per
i punteggi di salute mentale, mentre dall’altro la ricerca di validazione esterna può generare ansia da
prestazione.
Oltre alle dimensioni psicologiche, un mediatore biologico cruciale si è visto essere il sonno.
L’uso serale dei social media ritarda l’addormentamento e provoca attivazione cognitiva/fisiologica
(arousal) che peggiora la qualità del sonno. La deprivazione di sonno riduce la capacità di regolazione emotiva, aumentando la suscettibilità ad ansia e depressione il giorno successivo, che a loro volta
possono portare a un maggior uso di social media come coping maladattivo.
Anche i fattori sociodemografici sembrano essere rilevanti nella relazione tra dipendenza e salute
mentale.
Si è visto come le studentesse (16-25 anni) mostrino sistematicamente indicatori peggiori rispetto ai
ragazzi in quasi tutte le metriche considerate. Per quanto riguarda la salute mentale, la prevalenza
di sintomatologia ansiosa moderata/grave può raggiungere il 41% nelle ragazze e il 15% nei ragazzi.
Le donne, inoltre, ottengono punteggi BSMAS più elevati. Questo potrebbe essere attribuito a
diversi fattori: una maggiore tendenza relazionale che si traduce in un uso più intenso delle piattaforme
comunicative, una maggiore sensibilità al feedback sociale e una pressione socioculturale più forte
sull’immagine corporea, che i social media amplificano.
Si è rilevata anche un’influenza dello status socioeconomico. Per quanto riguarda la salute mentale,
studenti provenienti da contesti a basso reddito mostrano tassi più elevati di depressione e ansia.
Per quanto riguarda l’uso dei social media, l’uso problematico può essere più frequente in contesti svantaggiati a causa della mancanza di alternative ricreative offline e di una minore educazione digitale in
famiglia. A queste evidenze si aggiungono quelle sulla localizzazione geografica come determinante del
benessere psicologico. Gli studenti delle aree rurali mostrano spesso livelli più elevati di sintomi depressivi, stress e rischio suicidario rispetto ai pari urbani. Fattori contribuenti includono l’isolamento geografico, la minore accessibilità ai servizi di salute mentale, lo stigma sociale più marcato nelle piccole
comunità e le minori opportunità economiche/educative. In questo contesto, i social media possono
rappresentare un’ancora di salvezza per la connessione, ma anche un veicolo di confronto doloroso con
stili di vita urbani irraggiungibili. Tuttavia, gli studenti urbani affrontano stressor specifici come l’alta
densità abitativa, la competizione accademica e sociale, e il costo della vita.
Si evidenzia, infine, come la coesione familiare e uno stile genitoriale supportivo sono fattori protettivi robusti contro la dipendenza dai social media. Al contrario, conflitti familiari, mancanza
di supporto emotivo e scarsa supervisione sono predittori significativi dell’uso problematico dei social
media, spesso utilizzato come meccanismo di fuga da un ambiente domestico stressante.

**3.1. Descrizione del dataset**

La presente ricerca si basa sul dataset “Student Social Media & Relationships”, ottenuto da Kaggle.
Si tratta di una raccolta di dati cross-sectional ottenuta tramite un sondaggio online somministrato
nel primo trimestre del 2025. Il dataset offre una prospettiva globale, includendo risposte di studenti
provenienti da diversi contesti geografici (tra cui USA, Brasile, Germania, India e Corea del Sud) e
diversi livelli accademici (Scuola Superiore, Laurea Triennale/Magistrale e Dottorato).
Il dataset è composto da 705 osservazioni e le seguenti variabili:
* Student ID: Un identificativo intero univoco assegnato a ciascun intervistato per consentire la
deduplicazione e tracciare i singoli record senza rivelare informazioni personali.
* Age: Età dello studente in anni compiuti al momento dell’indagine.
* Gender: Il genere auto-dichiarato dallo studente, registrato come “Maschio” o “Femmina”.
* Academic Level: Il livello di istruzione più alto a cui è attualmente iscritto l’intervistato, con le
categorie: “Scuola superiore”, “Laurea triennale” o “Laurea magistrale”.
* Country: Il paese di residenza in cui lo studente ha completato l’indagine.
* Avg Daily Usage Hours: Il numero medio di ore al giorno che lo studente trascorre sulle piattaforme dei social media, calcolato in base alle stime di utilizzo auto-dichiarate nei giorni
feriali e nei fine settimana.
* Most Used Platform: La piattaforma di social media su cui lo studente trascorre più tempo,
include Facebook, Instagram, KakaoTalk, LINE, LinkedIn, Snapchat, TikTok, Twitter, VKontakte, WeChat, Whatsapp e Youtube.
* Affects Academic Performance: Un indicatore binario (“Sì”/“No”) che indica se lo studente
percepisce l’utilizzo dei social media come un impatto negativo sul proprio rendimento scolastico.
* Sleep Hours Per Night: Durata media del sonno notturno dell’intervistato in ore, fornita per
indagare le correlazioni tra tempo trascorso davanti allo schermo e qualità/quantità del sonno.
* Mental Health Score: Un numero intero autovalutato da 1 (scarso) a 10 (eccellente) che indica il
benessere mentale generale, consentendo di valutare potenziali associazioni con le abitudini sui
social media.
* Relationship Status: L’attuale stato sentimentale dello studente, classificato come “Single”, “In
Relazione” o “Complicato”.
* Conflicts Over Social Media: Il numero di discussioni o disaccordi che lo studente dichiara di
aver avuto con familiari, amici o partner a causa dell’uso dei social media, funge da indicatore
dell’attrito sociale.
* Addicted Score: Un punteggio composito da 1 (bassa dipendenza) a 10 (alta dipendenza) basato
su una scala di indagine standardizzata (ad esempio, la Bergen Social Media Addiction Scale),
che quantifica il grado di utilizzo problematico.

Poiché il dataset presenta diverse variabili categoriali, effettuiamo encoding. Si è scelto di eliminare le variabili non informative (Student_ID) e le variabili ad alta cardinalità (Country) perché la loro inclusione avrebbe reso il modello poco efficiente. Le variabili nominali come Gender, Most_Used_Platform, Affects_Academic_Performance e Relationship_Status sono state convertite in fattori, in modo che R sappia trattare correttamente le categorie durante la regressione. La variabile Academic_Level, in quanto variabile qualitativa ordinale, è stata trasformata in un fattore che tenga conto della gerarchia delle modalità.

```r
cols_to_remove <- c("Student_ID", "Country")
Data_Model <- Data[, !(names(Data) %in% cols_to_remove)]

cols_to_factor <- c("Gender", "Most_Used_Platform",
                    "Affects_Academic_Performance", "Relationship_Status")
Data_Model[cols_to_factor] <- lapply(Data_Model[cols_to_factor], as.factor)


Data_Model$Academic_Level <- factor(Data_Model$Academic_Level, levels = c("High School", "Undergraduate", "Graduate"), ordered = TRUE)
```

Questo processo è essenziale per permettere le analisi successive. La regressione lineare, infatti, richiede che le categorie siano trasformate in valori numerici; i metodi di selezione delle variabili (Stepwise e Boruta), che pur gestiscono bene i fattori, beneficiano dell’encoding nella definizione delle categorie e delle feature informative.

**3.3.1. Modello di Regressione Lineare Multipla**

È stato implementato un modello di regressione lineare multipla per valutare la capacità predittiva delle variabili indipendenti. Al fine di garantire la stabilità dei coefficienti e prevenire fenomeni di multicollinearità, sono state escluse le variabili che presentavano una correlazione eccessiva con il target (Addicted_Score, Conflicts_Over_Social_Media, Affects_Academic_Performance), evitando così distorsioni nelle stime degli altri predittori.

```r
Data_Model <- Data_Model %>%  mutate(across(where(is.numeric)
                              & !all_of("Mental_Health_Score"), scale))

Data_Final <- Data_Model %>% select(-Addicted_Score, -Conflicts_Over_Social_Media,                                     -Affects_Academic_Performance)

modello_lineare <- lm(Mental_Health_Score ~ ., data = Data_Final)
summary(modello_lineare)
```

Il modello ha mostrato un’ottima capacità predittiva, con un $R^2$ del 71.7%. Dall’interpretazione
dei coefficienti emerge un quadro chiaro:
* Il tempo trascorso online è il predittore più forte e negativo (β ≈ −0.76). Nello specifico, le
piattaforme di intrattenimento visivo (TikTok, Instagram) risultano avere un impatto negativo,
a differenza di quelle di messaggistica (Whatsapp) che mostrano un impatto positivo.
* Il sonno influisce positivamente sulla salute mentale, sebbene il suo effetto non riesca a compensare completamente l’impatto dell’uso intensivo.

**3.3.2. Modello Lineare Generalizzato**

La regressione lineare multipla assume che la variabile target sia continua e distribuita normalmente,
con un range che può spaziare da meno a più infinito. Al contrario, i punteggi del nostro dataset sono
valori discreti limitati (con range 1-10) e non negativi, si tratta quindi di dati di conteggio.
Per rispettare questi vincoli strutturali, si è quindi optato per la distribuzione di Poisson. Questa
scelta garantisce che le previsioni del modello siano confinate entro valori strettamente positivi, rispecchiando la natura della variabile target.
Tuttavia, l’applicazione di un modello di Poisson standard richiede il presupposto di “equidispersione”, ovvero l’uguaglianza tra media e varianza ($E(Y ) = V ar(Y )$). Un’analisi preliminare ha rivelato
una violazione di questo assunto, evidenziando una marcata sottodispersione: il parametro di dispersione calcolato (0.088) indica che la varianza effettiva dei punteggi è sensibilmente inferiore alla loro media. Utilizzare una Poisson classica in questo contesto avrebbe prodotto errori standard distorti e
p-value non attendibili. Per correggere tale distorsione, si è scelto di adottare il modello Quasi-Poisson,
che svincola il calcolo dal rigido presupposto di equidispersione, adattando le stime alla variabilità reale
del campione per ottenere risultati statisticamente robusti.
A completamento della struttura del modello, è stata integrata una funzione di collegamento logaritmica. Questa funzione ci ha permesso di evitare punteggi negativi e di interpretare i risultati in
termini moltiplicativi. Invece di limitarci a osservare variazioni puntuali, questa funzione consente di
quantificare l’impatto dei predittori in termini di variazioni percentuali.
Per la costruzione del modello lineare generalizzato le variabili indipendenti verranno standardizzate
e verranno escluse le variabili con elevata correlazione. Da un’analisi preliminare le variabili Age e Relationship_Status sono risultate non significative, di conseguenza sono state scartate.

```r
Modello_GLM <- glm(Mental_Health_Score ~ Gender + Avg_Daily_Usage_Hours +
                   Sleep_Hours_Per_Night + Academic_Level  + Most_Used_Platform,
                   family = quasipoisson(link = "log"),
                   data = Data_Final)
summary(Modello_GLM)
```

I risultati
ottenuti sono:
* Intercetta ($β_0 = 1.842$): rappresenta il valore atteso del logaritmo del punteggio di salute mentale
quando tutte le variabili numeriche sono a zero e tutte le variabili categoriche sono al loro livello
di riferimento. Il valore predetto del punteggio è $e^1.842 ≈ 6.31$.
* La variabile GenderMale ($β = 0.024$; $e^0.024 ≈ 0.976$) mostra come gli studenti abbiano una
riduzione del 2.4% del punteggio di salute mentale rispetto alle studentesse, a parità di ogni altra
variabile. L’effetto è piccolo ma statisticamente significativo.
* La variabile Avg Daily Usage Hours ($β = 0.127$; $e^0.127 ≈ 0.881$) si dimostra essere il predittore
più critico. Per ogni ora extra passata sui social, il punteggio di salute mentale diminuisce del
11.9%. è un effetto cumulativo: due ore extra riducono il punteggio del 22.4% (0.8812).
* La variabile Sleep Hours Per Night ($β = 0.021$; $e^0.021 ≈ 1.021$) presenta un effetto protettivo
sulla salute mentale: ogni ora di sonno in più aumenta il punteggio del 2.1%.
* Per la variabile Academic Level possiamo osservare sia il trend lineare sia il trend quadratico:
  1. Academic Level.L (Trend Lineare) $β = 0.050$; $e^0.050 ≈ 1.051$: Esiste una relazione di proporzionalità diretta significativa. In generale, man mano che si sale nella scala dell’istruzione, il punteggio di salute mentale tende a salire. Ogni “gradino” accademico moltiplica il benessere per circa 1.05.
  2. Academic Level.Q (Trend Quadratico) $β = 0.034$; $e^−0.034 ≈ 0.967$: Questo valore indica la ”curvatura” della relazione. Essendo negativo, siamo di fronte a una parabola con la concavità rivolta verso il basso.

In base ai risultati ottenuti possiamo dividere la variabile Most Used Platform in 3 categorie, differenziate dall’effetto sul punteggio di salute mentale:

1. Piattaforme con effetto negativo: KakaoTalk ($β = 0.101$) e Snapchat ($β = 0.085$) hanno l’effetto peggiore. Usare KakaoTalk riduce il punteggio del 9.7% ($e^−0.101 ≈ 0.93$), mentre Snapchat dell’8.2% ($e^−0.085 ≈ 0.918$). Spesso queste app sono legate a una comunicazione istantanea e continua che può generare ansia da prestazione o dipendenza. TikTok ($β = 0.064$) e Instagram ($β = 0.062$): entrambe riducono il benessere di circa il 6% (rispettivamente $e^−0.064 ≈ 0.937$ e $e^−0.062 ≈ 0.939$). è interessante notare come siano quasi identiche: entrambe si basano su contenuti visivi altamente rifiniti (algoritmi di scroll infinito) che favoriscono il confronto sociale verso l’alto.

2. Piattaforme con effetto positivo: X, precedentemente Twitter, ($β = 0.052$) è associato a un aumento del 5.4% ($e^0.52 ≈ 1.054$) del punteggio. Questo potrebbe essere dovuto al fatto che questa piattaforma è più orientata al testo, alla condivisione di contenuti artistici e all’informazione. WhatsApp ($β = 0.039$) aumenta il punteggio del 4% ($e^0.039 ≈ 1.04$). Essendo uno strumento di messaggistica diretta, favorisce il mantenimento di relazioni reali e il supporto sociale, agendo come fattore protettivo invece che stressogeno.

3. Piattaforme non significative: per piattaforme come LinkedIn, YouTube, WeChat, LINE e VKontakte, i p-value sono superiori alla soglia di significatività ($p > 0.05$), ciò dimostra che non hanno

Oltre alla stima dei singoli coefficienti, l’affidabilità del modello è stata valutata attraverso l’analisi dei parametri di diagnostica finale.
Il Parametro di Dispersione, utilizzato per verificare la relazione tra media e varianza e per convalidare la robustezza degli errori standard calcolati, in questo caso è 0.057. Ciò indica sottodispersione: i punteggi di salute mentale nel dataset sono molto “compatti” e vicini tra loro rispetto a quanto ci si
aspetterebbe. Usando la famiglia quasipoisson, il modello ha rimpicciolito gli Standard Error, rendendo i p-value più affidabili.
Il modello presenta inoltre una Null deviance (ovvero l’errore del modello che usa solo l`ıintercetta
per indovinare il punteggio) di 139.286 on 704 DF, e una Residual deviance (ovvero l’errore che rimane
dopo aver aggiunto i predittori) di 40.327 on 688 DF. Notiamo che l’errore viene ridotto del 71%.
Questo indica che le variabili indipendenti sono eccellenti predittori della salute mentale.

**3.4.2. Conclusioni finali sul punteggio di Dipendenza**

Dall’analisi dei punteggi di dipendenza emerge un quadro altrettanto nitido e coerente: la perdita
di controllo digitale non è un evento casuale, ma il risultato diretto di come lo studente decide di
abitare lo spazio virtuale. Il modello ottimizzato è riuscito a spiegare oltre il 91% di questo fenomeno,
una precisione statistica eccezionale che ci permette di identificare con estrema chiarezza i motori del
comportamento compulsivo.
Il dato più significativo è la conferma che la dipendenza non è solo una questione di tempo, ma di
impatto reale sulla vita. Il modello evidenzia come il principale campanello d’allarme sia il conflitto
con il rendimento accademico: la dipendenza si manifesta con forza quando lo studente percepisce che
il digitale sta erodendo i suoi obiettivi concreti. Non è dunque il “troppo utilizzo” in sé a definire il
problema, ma la sua capacità di interferire con i doveri e le ambizioni del quotidiano.
Proprio come per la salute mentale, assistiamo a una netta gerarchia del rischio tra le piattaforme.
YouTube si staglia come l’ambiente più insidioso: la sua architettura basata su flussi video continui e
algoritmi di raccomandazione sembra creare un “loop” di consumo passivo difficile da interrompere. Al
contrario, le app di messaggistica pura come WhatsApp, LINE e KakaoTalk agiscono quasi come fattori
di protezione. Questo ci suggerisce che quando il social media rimane uno strumento di comunicazione
diretta e finalizzata, il rischio di sviluppare un legame patologico crolla drasticamente rispetto ai social
basati sull’intrattenimento infinito.
In questo scenario, il sonno si conferma ancora una volta la difesa più preziosa. I risultati mostrano
che un buon riposo notturno non è solo un effetto collaterale di un minor uso dello smartphone, ma una
risorsa attiva: ogni ora di sonno guadagnata agisce come un freno alla vulnerabilità digitale, fornendo
allo studente le risorse cognitive necessarie per regolare i propri impulsi e mantenere un rapporto sano con la tecnologia.

**4. Conclusioni**

L’intera analisi delinea un panorama digitale in cui la salute mentale e la dipendenza degli studenti
non sono conseguenze inevitabili della tecnologia, ma l’esito di un’interazione specifica tra l’architettura
delle piattaforme e i ritmi biologici individuali. Il confronto tra i due modelli rivela una convergenza
quasi totale, con un’unica, fondamentale differenza nel ruolo di alcune piattaforme.
In entrambi i fenomeni osservati, il sonno emerge come la variabile più stabile e protettiva. In
un sistema complesso dove le app competono per l’attenzione, ogni ora di riposo notturno agisce
contemporaneamente come un rigeneratore dell’umore e come un freno agli impulsi compulsivi. è la
dimostrazione che la salute digitale non si costruisce solo “online”, ma soprattutto salvaguardando i
bisogni primari “offline”.
L’aspetto più rivelatore emerge confrontando come le diverse piattaforme influenzano i due modelli.
Sebbene l’impatto complessivo sia simile, le sfumature d’uso mostrano meccanismi psicologici differenti:
* Piattaforme di Dipendenza “Algoritmica” (YouTube): Mentre YouTube non è risultato il principale colpevole del calo di salute mentale, esso domina il modello della dipendenza. Questo
suggerisce che il suo formato video lungo e il sistema di raccomandazioni creino un legame
puramente compulsivo, che sequestra il tempo dello studente senza necessariamente generare,
nell’immediato, lo stress sociale tipico di altri network.
* Piattaforme di “Confronto Sociale” (TikTok, Instagram, Snapchat): Questi social si sono confermati i principali driver del calo di salute mentale. Qui il problema non è solo il tempo speso, ma la
natura del contenuto: l’esposizione costante a standard estetici elevati e la cultura dell’immagine
istantanea creano un “rumore ansiogeno” che erode il benessere psicologico, indipendentemente
dal fatto che l’utente si senta o meno tecnicamente “dipendente”.
* Piattaforme di Utilità (WhatsApp, LINE, Twitter/X): In entrambi i modelli, queste app occupano una posizione neutrale o addirittura positiva. La loro natura di strumenti per la comunicazione diretta o per l’aggiornamento informativo impedisce la creazione di quei “loop” ipnotici
tipici delle piattaforme basate sullo scrolling infinito.
